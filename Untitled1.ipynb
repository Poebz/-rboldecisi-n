## Arboles de decisi√≥n - EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv("winequality-red.csv", sep=',')  
print(df.columns.tolist()) 
print(df.head())
print(df.describe())
print(df.info())
print(df.columns)

# Distribuci√≥n de la variable objetivo
sns.countplot(x='quality', data=df)
plt.title("Distribuci√≥n de la calidad del vino")
plt.show()

# Matriz de correlaci√≥n
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de correlaci√≥n")
plt.show()

```

## Preprocesamiento

```python
# Verificar valores nulos
print(df.isnull().sum())

# (Opcional) Agrupar calidad en clases: baja (3-5), media (6), alta (7-8)
df['quality_label'] = df['quality'].apply(
    lambda x: 'baja' if x <= 5 else ('media' if x == 6 else 'alta')
)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['quality_label'] = le.fit_transform(df['quality_label'])
```

## Selecci√≥n de caracter√≠sticas

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = df.drop(['quality', 'quality_label'], axis=1)
y = df['quality_label']

selector = SelectKBest(score_func=f_classif, k=8)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
print("Caracter√≠sticas seleccionadas:", selected_features)
```

## Divisi√≥n en train y test

```python
from sklearn.model_selection import train_test_split

X = df[selected_features]
y = df['quality_label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.tree import DecisionTreeClassifier

# Crear el modelo
model = DecisionTreeClassifier(random_state=42, max_depth=5)  # Puedes ajustar los hiperpar√°metros

# Entrenar el modelo
model.fit(X_train, y_train)

```

## Entrenamiento del modelo

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred, target_names=le.classes_))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
disp.plot(cmap='Blues')
plt.show()
```

## Visualizaci√≥n

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(model, feature_names=selected_features, class_names=le.classes_, filled=True)
plt.title("√Årbol de Decisi√≥n")
plt.show()
```
